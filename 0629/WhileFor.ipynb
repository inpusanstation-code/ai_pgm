```python
# range (start, stop, step) : start부터 stop까지 step만큼 증가시키면서 반복

#1부터 100까지의 합을 출력하는 프로그램
total = 0
for i in range(1, 101):
    total += i
print(f"1부터 100까지의 합: {total}")

#1부터 100까지의 수 중에 홀수의 합을 출력하는 프로그램
odd_total = 0
for i in range(1, 101):
    if i % 2 == 1:
        odd_total += i
print(f"1부터 100까지의 홀수 합: {odd_total}")

#1부터 100까지의 수 중에 6의 배수들의 합을 출력하는 프로그램
multiple_of_6_total = 0
for i in range(1, 101):
    if i % 6 == 0:
        multiple_of_6_total += i
print(f"1부터 100까지의 6의 배수 합: {multiple_of_6_total}")



```

```python
number=int(input("Enter a number: "))
for i in range(1, number + 1):
    for j in range(1, i + 1):
        print("*", end="")
    print()  # 줄바꿈
```

```python
#빗변이 왼쪽에 있는 직각삼각형
number=int(input("Enter a number: "))
for i in range(1, number + 1):
    for j in range(1, number - i + 1):
        print(" ", end="")
    for k in range(1, i + 1):
        print("*", end="")
    print()  # 줄바꿈
    
```

```python
#정삼각형 별찍기 
number=int(input("Enter a number: "))
for i in range(1, number + 1):
    for j in range(1, number - i + 1):
        print(" ", end="")
    for k in range(1, 2 * i):
        print("*", end="")
    print()  # 줄바꿈
    
```

```python
#마름모 별찍기
number=int(input("Enter a number: "))  

for i in range(1, number + 1):
    for j in range(1, number - i + 1):
        print(" ", end="")
    for k in range(1, 2 * i):
        print("*", end="")
    print()

for i in range(number - 1, 0, -1):
    for j in range(1, number - i + 1):
        print(" ", end="")
    for k in range(1, 2 * i):
        print("*", end="")
    print()
```
