```python
#팩토리얼 코드
n = int(input("Enter a non-negative integer: "))
if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = 1
    for i in range(1, n + 1):
        result *= i
    print(f"The factorial of {n} is {result}.")
    
```

```python
#섭씨, 화씨 변환기
import decimal

def round_decimal(value):
    return value.quantize(decimal.Decimal("0.01"), rounding=decimal.ROUND_HALF_UP)

def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit."""
    return (celsius * decimal.Decimal("9") / decimal.Decimal("5")) + decimal.Decimal("32")

def fahrenheit_to_celsius(fahrenheit):
    """Convert Fahrenheit to Celsius."""
    return (fahrenheit - decimal.Decimal("32")) * decimal.Decimal("5") / decimal.Decimal("9")

def main():
    print("Temperature Converter")
    choice = input("Type 'C' to convert Celsius to Fahrenheit or 'F' to convert Fahrenheit to Celsius: ").strip().upper()
    if choice == 'C':
        celsius = decimal.Decimal(input("Enter temperature in Celsius: "))
        fahrenheit = celsius_to_fahrenheit(celsius)
        print(f"{celsius}°C is equal to {round_decimal(fahrenheit)}°F")
    elif choice == 'F':
        fahrenheit = decimal.Decimal(input("Enter temperature in Fahrenheit: "))
        celsius = fahrenheit_to_celsius(fahrenheit)
        print(f"{fahrenheit}°F is equal to {round_decimal(celsius)}°C")
    else:
        print("Invalid input. Please enter 'C' or 'F'.")

if __name__ == "__main__":

    main() 
```

```python
#Recursive factorial function
def factorial(n):
    """Calculate the factorial of a non-negative integer n."""
    if n == 0:
        return 1
    else: 
        return n * factorial(n - 1)  # Recursive call to calculate factorial
    
def main(): 
    number = int(input("Enter a number to calculate its factorial: "))
    print(f"The factorial of {number} is {factorial(number)}")

if __name__ == "__main__":
    main()

```

```python
#fibonacci sequence generator using recursion (bad for large n)
def fibonacci(n):
    """Return the nth Fibonacci number using recursion."""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)  
    
if __name__ == "__main__":
    count = 100  # Generate the first 10 Fibonacci numbers
    fib_sequence = [fibonacci(i) for i in range(count)]
    print(f"The first {count} Fibonacci numbers are: {fib_sequence}")
```
